```python
%matplotlib notebook
import pandas as pd
from pylab import *
```

```python
data = pd.read_excel("Known_set_Bacillus.xlsx")
data.head()
```

```python
y = data["PA"]
x = data[list(set(data.columns)-{"PA","UTR5","ORF"})]
```

```python
cm = pd.concat([x,y],axis = 1).corr()
imshow(cm.values-np.eye(cm.shape[0]))
xticks(np.arange(cm.shape[0]),cm.columns)
yticks(np.arange(cm.shape[0]),cm.columns)
colorbar()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)
print(reg.coef_)
```

```python
y_pred = reg.predict(x_test)
```
